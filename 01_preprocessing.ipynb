## 準備

```python
import os
import re

import pandas as pd
import numpy as np
```

```python
input_path = "data/input/"
```

## 前処理

```python
# ファイル読み込み
cols = ["原文"]
df = pd.read_csv(input_path+"rawdata.csv", header=None, names=cols, encoding="utf-8")
```

```python
# 原文前処理
def preprocessing(x):
    s = x.replace(" ","")

    ret = []
    ret.append(re.search(r"[0-9]*", s).group(0).zfill(5))   # 発表番号
    ret.append(re.search(r"\n(.+)\n", s).group(1))   # 発表タイトル
    org = re.findall("(?<=[(（]).*?(?=[)）])", s)[-1].replace("（研究委員会", "") # 発表組織
    org_cat = "大学" if "大学" in org else "企業" # 発表組織カテゴリ
    ret += [org, org_cat]   # 発表組織

    return pd.Series(ret)


df[["発表番号", "発表タイトル", "発表組織", "発表組織カテゴリ"]] = df["原文"].apply(preprocessing)
del df["原文"]
df = df[:-5]    # 特別講演は排除する
```
